Import pandas library

```python
import pandas as pd
```

Load a CSV file

```python
df1 = pd.read_csv("oecd_bli_2015.csv", thousands=',')
```

Display first 5 rows

```python
df1.head()
```

Find missing values in DataFrame and count them coulum-wise

```python
df1.isnull().sum()
```

Filter rows in a DataFrame where a column Value > 100

```python
df1[df1["Value"]>100]
```

Group a DataFrame by Country column and calculate the mean of each group

```python
country_mean = df1.groupby("Country")['Value'].mean()
country_mean
```

Sort a DataFrame by multiple rows (first by Country ascending, then by Inequality ascending and then by Value Descending)

```python
sorted_by_multi_cols = df1.sort_values(by=["Country","Inequality","Value"],ascending=[True,True,False])
sorted_by_multi_cols
```

Add a new column to a DataFrame that is transformation of existing columns. (Double_Value which is transformation of Value by multiplying with 2)

```python
df1["Double_of_Value"] = df1["Value"]*2
df1
```

Load another CSV file

```python
df2 = pd.read_csv("gdp_per_capita.csv", thousands=',')
```

Merge two DataFrames (df1 and df2) on Country column by Inner Join

```python
country_inner_merged = pd.merge(df1,df2,on="Country",how="inner")
country_inner_merged
```
